```python
import pandas as pd

abt = pd.read_csv('../../data/D2/abtclean.csv', sep='|')
buy = pd.read_csv('../../data/D2/buyclean.csv', sep='|')
gt = pd.read_csv('../../data/D2/gtclean.csv', sep='|')

abt.drop(columns=['description', 'price'], inplace=True)
buy.drop(columns=['description', 'price'], inplace=True)

abt['party'] = 'abt'
buy['party'] = 'buy'


for idx, row in gt.iterrows():
    abt_id = row['D1']
    buy_id = row['D2']
    abt.loc[abt['id'] == abt_id, 'group_id'] = idx
    buy.loc[buy['id'] == buy_id, 'group_id'] = idx

buy['id'] = buy['id'] + 10_000
abt.to_csv('abt_primat.csv', index=False, sep=',')
buy.to_csv('buy_primat.csv', index=False, sep=',')


concatenated = pd.concat([abt, buy], ignore_index=True)
concatenated.to_csv('primat.csv', index=False, sep=',')


abt.head(10)
```
