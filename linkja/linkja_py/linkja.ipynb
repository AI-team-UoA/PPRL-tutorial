## Load CSVs with Linkja

```python
from utils import load_csv_fields
import pandas as pd

abt = pd.read_csv("../../data/D2/abt.csv", sep="|")
buy = pd.read_csv("../../data/D2/buy.csv", sep="|")


abt.to_csv("abt.csv", index=False, sep=",")
buy.to_csv("buy.csv", index=False, sep=",")

abt_linkja = load_csv_fields("abt.csv", ["name"])
buy_linkja = load_csv_fields("buy.csv", ["name"])

abt_linkja[:10]
```

## Brute Force Matching

```python
from record_linker import compare_records

matches = []


for abt_rec in abt_linkja:
    for buy_rec in buy_linkja:
        match, score = compare_records(abt_rec,
                buy_rec,
                salt=" ", threshold=0.01, ngram=4)
        if match:
            matches.append((abt_rec, buy_rec, score))
            break

print(f"Total matches found: {len(matches)}")
```

```python
print(buy_linkja[0])
print(abt_linkja[1023])


match, score = compare_records(abt_linkja[1023], buy_linkja[0], salt=" ", threshold=0.6, ngram=4)
if match:
    print(f"Match: abt-1023 <--> buy-0 | Score: {score:.2f}")
else:
    print(f"No match: abt-1023 <--> buy-0 | Score: {score:.2f}")
```
